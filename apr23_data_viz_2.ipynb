```python
#collect only that data where publisher, platform and genre is among top 3 in terms of count
```

```python
df["Publisher"].value_counts()
```

```python
df["Publisher"].value_counts().index
```

```python
top3_pub= df["Publisher"].value_counts().index[:3]
```

```python
top3_platform = df["Platform"].value_counts().index[:3]
```

```python
top3_genre = df["Genre"].value_counts().index[:3]
```

```python
top3_data = df.loc[
    (df["Genre"].isin(top3_genre))
    & (df["Platform"].isin(top3_platform))
    & (df["Publisher"].isin(top3_pub))
    ]
```

```python
top3_data.head()
```

```python
sns.countplot(x = "Publisher", data = top3_data) #univariate analysis - cat
```

```python
#Bivariate--> Cat-Cat: For each publisher, for each platform
```

```python
sns.countplot(x = "Publisher", hue = "Platform", data = top3_data)
```

```python

plt.figure(figsize=(20,12)).suptitle("NA Sales vs regions",fontsize=20)
# Using a 2x3 subplot
plt.subplot(2, 3, 1)
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=top3_data)


plt.subplot(2, 3, 3)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top3_data, color='red')


plt.subplot(2, 3, 4)
sns.scatterplot(x='NA_Sales', y='Other_Sales', data=top3_data, color='green')

plt.subplot(2, 3, 6)
sns.scatterplot(x='NA_Sales', y='Global_Sales', data=top3_data, color='orange')

plt.show()
```
