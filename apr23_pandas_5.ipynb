```python
def temp_avg(df):
  df["Temp_avg"] = df["Temperature"].mean()

  return df
```

```python
data_tidy[data_tidy["Drug_Name"] == "ketamine hydrochloride"].head()
```

```python
data_tidy.head(10)
```

```python
data_tidy.head(20)
```

```python
temp_points = [5, 20, 35, 50, 60]
temp_lables = ["low", "medium", "high", "very_high"]
```

```python
data_tidy["temp_category"] = pd.cut(data_tidy["Temperature"],
                                    bins = temp_points,
                                    labels = temp_lables)
```
